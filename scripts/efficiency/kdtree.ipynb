```python
from scipy import spatial
import numpy as np
import json
import os
dataset='assistment'
stg='MFI'
post_fix='30'
path_prefix = os.path.abspath('../../')
trait = json.load(open(f'{path_prefix}/data/{dataset}/{stg}/ball_trait{post_fix}.json', 'r'))
vec_v=[]
for i in trait:
    vec_v.append(np.linalg.norm(np.array(i)))
max_v = max(vec_v)
new_trait=[]
for i in trait:
    extra_v = np.sqrt(max_v**2-np.linalg.norm(np.array(i))**2)
    # print(extra_v)
    tmp=[extra_v]
    tmp.extend(i)
    new_trait.append(tmp)
query = json.load(open(f'{path_prefix}/data/{dataset}/{stg}/query.json', 'r'))
tree = spatial.KDTree(new_trait)
# tree.data
```

```python

import time
starttime= time.time()
times=[]
# pts = np.array([[0, 0], [2.1, 2.9]])
for qs in query:
    for it,q in enumerate(qs):
        s_t=time.time()
        tmp=[0]
        tmp.extend(q)
        tree.query(tmp,it+1)
        # u.get_nns_by_vector(q, it+1, search_k=-1, include_distances=False) 
        times.append(time.time()-s_t)
endtime= time.time()
print((endtime-starttime)/len(times)*10)
print(max(times))

```

```python
import numpy as np
import pandas as pd
# len(times)
s_time=[]
tmp=[]
for i,t in enumerate(times):
    tmp.append(t)
    if len(tmp)==10:
        s_time.append(tmp)
        tmp=[]
s_time=np.array(s_time)
s_time = np.average(s_time, axis=0, weights=None, returned=False)
s_time
# s_time/29*47
res=[]
a_time=0
for t in s_time:
    # print(t)
    # break
    a_time+=t
    res.append(a_time)
# print(s_time.shape)
df = pd.DataFrame(res)
df = df.transpose()

df.to_csv(f'data/kdtree_{dataset}_{stg}.csv')
```
